```python
import os
import numpy as np
txt_files = [f for f in os.listdir('.') if f.endswith('.txt') and os.path.isfile(f)]

# list of province in indonesia
province = np.array([str(f)[:-4] for f in txt_files])

print('Number of province:', len(province))
print(province)
```

```python
import pandas as pd 


production_data = {}

for p in province:
    ds = pd.read_excel("Data Kebutuhan Listrik.xlsx", sheet_name=p)
    ds = ds.iloc[:-1,:]
    ds['Tahun'] = np.int64(ds['Tahun'])
    ds = ds[(ds['Tahun'] >= 2025) & (ds['Tahun'] <= 2034)]

    data = ds['Produksi(GWh)']
    tahun = ds['Tahun']

    production_data[p] = dict(zip(tahun, data))

    df = pd.DataFrame(
    production_data
)

df = df.transpose()
df.index.name = 'Province'
df
```

```python
region_map = {
    'Aceh': 'Sumatera',
    'Bali': 'Jawa-Bali',
    'BangkaBelitung': 'Sumatera',
    'Banten': 'Jawa-Bali',
    'Bengkulu': 'Sumatera',
    'DIY': 'Jawa-Bali',
    'Gorontalo': 'Sulawesi',
    'Jakarta': 'Jawa-Bali',
    'Jambi': 'Sumatera',
    'JawaBarat': 'Jawa-Bali',
    'JawaTengah': 'Jawa-Bali',
    'JawaTimur': 'Jawa-Bali',
    'KalimantanBarat': 'Kalimantan',
    'KalimantanSelatan': 'Kalimantan',
    'KalimantanTengah': 'Kalimantan',
    'KalimantanTimur': 'Kalimantan',
    'KalimantanUtara': 'Kalimantan',
    'KepulauanRiau': 'Sumatera',
    'Lampung': 'Sumatera',
    'Maluku': 'Maluku',
    'MalukuUtara': 'Maluku',
    'NTB': 'NTB',
    'NTT': 'NTT',
    'Papua': 'Papua',
    'PapuaBarat': 'Papua',
    'Riau': 'Sumatera',
    'SulawesiBarat': 'Sulawesi',
    'SulawesiSelatan': 'Sulawesi',
    'SulawesiTengah': 'Sulawesi',
    'SulawesiTenggara': 'Sulawesi',
    'SulawesiUtara': 'Sulawesi',
    'SumateraBarat': 'Sumatera',
    'SumateraSelatan': 'Sumatera',
    'SumateraUtara': 'Sumatera'
}
```

```python
df_with_region = df.copy()
df_with_region['Region'] = [region_map[prov] for prov in df_with_region.index]
cols = ['Region'] + [col for col in df_with_region.columns if col != 'Region']
df_with_region = df_with_region[cols]
df_with_region.sort_values(by='Region')
```

```python
gwh_per_region = df_with_region.groupby('Region').sum()
gwh_per_region
```

```python
gwh_per_region.to_excel(r'Rekapitulasi Data Demand.xlsx')
```

```python
import matplotlib.pyplot as plt 
from matplotlib.offsetbox import AnchoredText

fig, ax = plt.subplots(figsize=(10, 8), ncols=2, nrows=4, dpi=300, constrained_layout=True)
ax = ax.flatten()

for i in range(8):
    data = gwh_per_region.iloc[i]
    
    ax[i].bar(data.keys(), data.values / 1e3, zorder=2, color='tab:blue')
    
    ax[i].set_title(gwh_per_region.index[i], loc='left', fontweight='bold', fontsize=10)
    
    # at = AnchoredText(gwh_per_region.index[i],
    #                   loc='upper left', prop=dict(size=8), frameon=True,
    #                   )
    # ax[i].add_artist(at)
    ax[i].grid(which='major', axis='y', zorder=0, color='k')
    ax[i].set_ylabel(r'$10^{3}$ GWh')

    
fig.suptitle('Proyeksi Produksi Listrik Berdasarkan RUPTL', fontweight='bold')
plt.show()
```
